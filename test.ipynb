```python
import pandas as pd
import json

# Load the Excel file (replace with your filename)
df = pd.read_excel("data.xlsx")
```

```python
df = df.drop(0)
df.columns = df.iloc[0]
df = df.drop(df.index[0]).reset_index(drop=True)
```

```python
df = df.dropna()
```

```python
df
```

```python
df_melted = df.melt(id_vars=["Region, subregion, country or area"], var_name="key", value_name="value")

result = df_melted.to_dict(orient="records")

js_data = [
    {"group": row["Region, subregion, country or area"], "key": row["key"], "value": row["value"]}
    for row in result
]

with open("data.js", "w") as f:
    f.write("export default ")
    json.dump(js_data, f, indent=2, ensure_ascii=False)
    f.write(";")

print(js_data)
```
